### ONE DIMENSIONAL DATA SETS


Most basic function is pyplot.plt()

This takes two sets of numbers:
x-values (list or array)
y-values (list or array)

```python
# plot a bunch of random numbers

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.clf()

np.random.seed(1000)

y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x,y)

plt.title('Random numbers')
```

```python
# Note that in this case, even if you don't pass an x-value, 
# matplotlib is clever enough to handle that
plt.clf()

plt.clf()

np.random.seed(1000)

y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x,y)

plt.title('Without Specifying X')
```

```python
# You can also pass in a function

plt.clf()

plt.clf()

np.random.seed(1000)

y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x,y.cumsum())

plt.title('Passing in a function')
```

```python
# Turning off the grid, forcing equal scaling on both axes
plt.grid(False)
plt.axis('equal')

# Could also pass in:  empty, off, scaled, tight, image, or [xmin,ymin,xmax,ymax]

plt.clf()

np.random.seed(1000)

y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x,y.cumsum())

plt.title('No Grid; Equal Axes')
```

```python
# Explicitly setting min and max.  
# Here, we'll cut off everything above 0
plt.clf()

y = np.random.standard_normal(20)

plt.plot(x, y)

plt.title('Setting axis limits')
plt.xlim(-1, 20) # specifying upper and lower limit
plt.ylim(-0) # lower only
```

```python
# Using Labels + making the line look nice

plt.clf()

y = np.random.standard_normal(20)

#Plot with blue double dash:
plt.plot(x, y.cumsum(), 'b--', lw = 1.5, )

#and the points themselves in red dots
plt.plot(y.cumsum(), 'ro', lw=1.5)

plt.xlabel('Hello I am the x label')
plt.ylabel('Hello I am the y label')
plt.title('Using labels + making the line look nice')
```

### TWO DIMENSIONAL DATA SETS

```python
plt.clf()

x = np.arange(20)

i = 0
while i < 20:
    y = np.random.standard_normal(20)
    i += 1
    plt.plot(x, y.cumsum())

plt.title('Multiple data sets')
```

```python
# same thing, just 2d this time, slightly diff approach
plt.clf()

y = np.random.standard_normal((20,2)).cumsum(axis=0)

plt.title('Multiple data sets')

plt.plot(y, lw=1.5)
plt.plot(y, 'ro')
plt.xlabel('index')
plt.ylabel('value')
```

```python
#setting the legend, specifying labels

y = np.random.standard_normal((20,2)).cumsum(axis=0)

#Firstly, the first two series separately:

#1st one
plt.plot(y[:,0], 'b--', lw=1.5, label='1st')
#2nd
plt.plot(y[:,1], 'g--', lw=1.5, label='1st')

#Now just the dots themselves.
#Here we plot both simultaneously
plt.plot(y, 'ro')

#Plot legend.  Location 0 is 'best fit'
plt.legend(loc=0)

plt.title('Legend; Labels; Two series')
```

```python
# Now, different scales on the axis

y = np.random.standard_normal((20,2)).cumsum(axis=0)

plt.clf()

# Multiply the first data series by 100, so it's a diff scale
y[:,0] *= 100

#now we need to do the fig, ax stuff
fig, ax1 = plt.subplots()

# do the first one:
plt.plot(y[:,0], 'b--', lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')

plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('First Series y-value')

#Create a second y axis that shares the same x axis as ax1
ax2 = ax1.twinx()

plt.plot(y[:,1], 'g-', lw=1.5, label='2nd')
plt.plot(y[:,1], 'r*')
plt.ylabel('Second Series y-value')

#looks like you need to do each legend separately
plt.legend(loc=0)

plt.title('Two series; different axes')
```

```python
# Same data, two subplots instead

# Now, different scales on the axis

y = np.random.standard_normal((20,2)).cumsum(axis=0)

plt.clf()

# Multiply the first data series by 100, so it's a diff scale
y[:,0] *= 100

#now we need to do the fig, ax stuff
fig, ax1 = plt.subplots()

### NOW PUT ON SEPARATE PLOTS INSTEAD
plt.subplot(211) # 2 rows.  this is column 1, row 1

# do the first one:
plt.plot(y[:,0], 'b--', lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')

plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('First Series y-value')

#The placement of this line of code seems to be important
plt.title('Two series; different axes')


### NOW PUT ON SEPARATE PLOTS INSTEAD
plt.subplot(212) # 2 rows.  this is column 1, row 2


plt.plot(y[:,1], 'g-', lw=1.5, label='2nd')
plt.plot(y[:,1], 'r*')
plt.ylabel('Second Series y-value')

#looks like you need to do each legend separately
plt.legend(loc=0)
```

```python
# Bar chart for the 2nd series instead.
# Same code, one line difference
# And now we're doing side by side instead

y = np.random.standard_normal((20,2)).cumsum(axis=0)

plt.clf()

# Multiply the first data series by 100, so it's a diff scale
y[:,0] *= 100

#now we need to do the fig, ax stuff
fig, ax1 = plt.subplots()

### NOW PUT ON SEPARATE PLOTS INSTEAD
plt.subplot(121) # 2 rows.  this is column 1, row 1

# do the first one:
plt.plot(y[:,0], 'b--', lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')

plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('First Series y-value')

#The placement of this line of code seems to be important
plt.title('Two series; different axes')

### NOW PUT ON SEPARATE PLOTS INSTEAD
plt.subplot(122) # 2 rows.  this is column 1, row 2

plt.bar(np.arange(len(y)), y[:,1],  width=0.5, color='g', label='2nd')

#plt.plot(y[:,1], 'g-', lw=1.5, label='2nd')
plt.plot(y[:,1], 'r*')
plt.ylabel('Second Series y-value')

#looks like you need to do each legend separately
plt.legend(loc=0)
```

## Pyplot, + multiple charts on one sheet

```python
#Pyplot (wrqpper on top of matplotlib)
#Multiple Charts on one sheet

import pandas as pd
import matplotlib.pyplot as plt

#Sample data
from sklearn.datasets import load_boston
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
#boston['MEDV'] = boston_dataset.target


plt.figure(figsize=(40,50))
features = boston_dataset['feature_names']
target = boston_dataset.target

features = boston_dataset['feature_names']

#target = boston['MEDV']

for i, col in enumerate(features):
    plt.subplot(features.size/2, 3, i+1)
    x = boston[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')

    #Graph is taking up too much space - exit early
    if (i == 5):
        break

plt.show()
```

## Seaborn + Correlation Matrix / Heatmap

```python
#Seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston

boston_dataset = load_boston()

boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
#print (boston.head())
boston['MEDV'] = boston_dataset.target

sns.set(rc = {'figure.figsize':(6, 4)})
sns.distplot(boston['MEDV'], bins = 30)
plt.show()
```

```python
#Seaborn Heatmap
#Relies on 'boston' object above and associated imports

sns.set(rc = {'figure.figsize':(10, 6)})
correlation_matrix = boston.corr().round(2)
sns.heatmap(data = correlation_matrix, annot = True)
plt.show()
```
